# Linear Regression using python

The goal of linear regression is to find a line that best fits a set of data points.
There are following steps
1. Randomly initialize parameters for hypothesis function
2. Compute mean squared error
3. Calculate the partial derivative
4. Update the parameters based on derative and learning rate
5. repeat 2 - 4 until error is minimized or in other words we reach global minima

```python
# creating dataset
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt

X, y = make_regression(n_samples=100, n_features=1, noise=0.4, bias = 50)
```

```python
# hypothesis function
def hypothesis(theta0, theta1, x):
    return theta0 + (theta1*x)
```

In the begining, we randomly initialize our parameters, which means we give some random values to the two parameters and get the random line

```python
def plotline(theta0, theta1, X, y):
    max_x = np.max(X) + 100
    min_x = np.min(X) - 100
    
    xplot = np.linspace(min_x, max_x, 1000)
    yplot = theta0+theta1*xplot
    
    plt.plot(xplot, yplot, color='#ff0000', label = 'Regression Line')
    plt.scatter(X, y)
    plt.axis([-10,10,0,200])
    plt.show()
```

```python
theta0 = np.random.rand()
theta1 = np.random.rand()

plotline(theta0, theta1, X, y)
```

The above line is clearly wrong, so we calculate the error of the line using cost function
1/2 sum(Ypred - Yactual)2

```python
def cost(theta0, theta1, X, y):
    costValue = 0
    for (xi, yi) in zip(X, y):
        costValue += 0.5 * ((hypothesis(theta0, theta1, xi)-yi)**2)
    return costValue
```

Now, we have a value of how wrong the function is we need to adjust the function to reduce the error

### Calculating derivatives

Our goal is to minimize the cost function so for this we set the partial derivatives of the cost function to 0
* theta0 = hypothesis - Y
* theta1 = (hypothesis - Y) X

```python
def derivatives(theta0, theta1, X, y):
    dtheta0 = 0
    dtheta1 = 0
    for (xi,yi) in zip(X, y):
        dtheta0 += hypothesis(theta0, theta1, xi) - yi
        dtheta1 += (hypothesis(theta0, theta1, xi)-yi)*xi
        
        dtheta0 /= len(X)
        dtheta1 /= len(X)
        
    return dtheta0, dtheta1
```

Upating the parameters based on the learning rate, using gradient descent appraoch

Wnew = Wold - learning_rate* d/dwold

```python
def updateParameters(theta0, theta1, X, y, alpha):
    dtheta0, dtheta1 = derivatives(theta0, theta1, X, y)
    theta0 = theta0 - (alpha * dtheta0)
    theta1 = theta1 - (alpha * dtheta1)
    
    return theta0, theta1
```

### Minimizing the cost function

Now, we repeat these steps, checking the error, finding the derivatives, and updatign the weights until the error is minimum

```python
def LinearRegression(X, y):
    theta0 = np.random.rand()
    theta1 = np.random.rand()
    
    for i in range(0, 1000):
        if i % 100 == 0:
            plotline(theta0, theta1, X, y)
            
        theta0, theta1 = updateParameters(theta0, theta1, X, y, 0.005)
```

```python
LinearRegression(X, y)
```
